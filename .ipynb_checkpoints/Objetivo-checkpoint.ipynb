### <center> 0. Imagenes-datasets:
    
    Se decidió que el programa clasificara números (del 0 al 9) y letras (abecedario de mayúsculas como minúsculas). Guarde en la carpeta validación (que se encuentra en la carpeta imágenes) las imágenes para la validación del algoritmo.
    
    Para generar las imágenes se puede guiar de los video tutoriales de la sección de Teams con nombre "Imagenes-datasets". Se requieren imágenes para MINIMO dos letras y dos números, es decir, si usted por ejemplo elije 1,3 como números y 'x','g' como letras, en la carpeta de números tiene que haber imágenes de los números 1 y 3, y para el caso de las letras tienen que generarse imágenes con las letras 'x' y 'g'. Mínimo 10 imágenes para cada carácter (tanto letras como números) por lo tanto tienen que estar mínimo 40 imágenes como dataset de aprendizaje. Si usted desea agregar más imágenes para que el comportamiento del modelo sea más preciso no hay ningún problema, de igual forma si se quieren tener más de 2 letras o números (si usted quiere probar con todo el abecedario y con todos los números ya es decisión suya). Como dataset de validación mínimamente tienen que haber 4 imágenes una para validar el funcionamiento del algoritmo con cada carácter (2 letras y 2 números).
